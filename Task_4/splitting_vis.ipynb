```python
import os
import numpy as np
import pandas as pd
from tqdm import tqdm
import uuid
```

```python
file = 'train_triplets.txt'
df = pd.read_csv(file, sep=' ', names=['A', 'B', 'C'],
                 dtype='str')
```

```python
a, count = np.unique(df['A'], return_counts=True)
print(a)
print(count)
```

```python
A = df.A.unique().tolist()
B = df.B.unique().tolist()
C = df.C.unique().tolist()
overlapAB = list(set(A).intersection(B))
print(overlapAB)
overlapAC = list(set(A).intersection(C))
print(overlapAC)
overlapBC = list(set(B).intersection(C))
print(overlapBC)
#There are overlaps between rows
```
